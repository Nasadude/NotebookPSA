### Creat first visualization using matplotlib.pyplot

```python
# Create data to generate a line plot
x = np.linspace(-10,10,150)
print(f"x={x}")
y = x**2
print(f"y={y}")
```

```python
# Create a line plot of a quadratic curve
plt.plot(x,y)
plt.show()
```

```python
# Create data to generate a line plot
x = np.linspace(-10,10,150)
#print(f"x={x}")
y = x**2
#print(f"y={y}")
plt.figure(figsize=(10,4))
plt.plot(x,y)
#plt.axis("off")
plt.grid()
plt.title("Visualizing the Quadraditic Data",color="maroon", weight="bold",fontsize=15)
plt.xlabel('x - created from linspace',fontsize=10, color='m')
plt.ylabel('y: squaring x', fontsize=10,color='m')
plt.show()
```
